```python
# low realが*****の場合
import codecs
import glob
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import cm

data_dir = "data/alpha1 1A/*.out"
paths = [p for p in glob.glob(data_dir)]
key_1 = "Counterpoise corrected energy"
key_2 = "BSSE energy"
key_3 = "sum of fragments"
key_4_5 = "complexation energy"

data_names = data_dir.split('/')[2:-1]
data_name = ""
for p in data_names:
    data_name = data_name + p
result_dir = data_dir.split('*.out')[0] + data_dir.split('*.out')[0].split('/')[1] + ".csv"
df_output = pd.DataFrame()

for path in paths:
    # file_nameから角度の情報を取る
    file_name = path.split('/')[-1]

    with codecs.open(path, "r", "UTF-8", "ignore") as file:
        df = pd.read_table(file)
        df = df.rename(columns={df.columns[0]: 'Col_1'})
        # initial_keyを含む行を探す
        key_1_df = df[df['Col_1'].str.contains(key_1, case=False)]
        Counterpoise_corrected_energy = float(list(key_1_df['Col_1'])[0].split("=")[-1])
        
        key_2_df = df[df['Col_1'].str.contains(key_2, case=False)]
        BSSE_e = float(list(key_2_df['Col_1'])[0].split("=")[-1])
        
        key_3_df = df[df['Col_1'].str.contains(key_3, case=False)]
        sum_of_fragments = float(list(key_3_df['Col_1'])[0].split("=")[-1])
        
        key_4_5_df = df[df['Col_1'].str.contains(key_4_5, case=False)]
        complexation_energy_raw = float(key_4_5_df['Col_1'].iloc[0].split("=")[-1].split("kcal/mole")[0])
        complexation_energy_corrected = float(key_4_5_df['Col_1'].iloc[1].split("=")[-1].split("kcal/mole")[0])
        
        df_E = pd.DataFrame(data=[Counterpoise_corrected_energy, 
                                  BSSE_e,
                                  sum_of_fragments,
                                  complexation_energy_raw,
                                  complexation_energy_corrected
                                 ],
                            columns=[file_name],
                            index=["Counterpoise corrected energy",
                                  "BSSE energy",
                                   "sum of fragments",
                                   "complexation energy (raw)",
                                   "complexation energy (corrected)"
                                  ]).T
        df_output = pd.concat([df_output, df_E])
        
        
df_output.to_csv(result_dir)
df_output
```
